```python
#W=Rv+Cm/v+m
```

```python
R = movies_data['vote_average']
v = movies_data['vote_count']
C = movies_data['vote_average'].mean()
m = movies_data['vote_count'].quantile(0.7)
```

```python
movies_data['Weighted_Average'] = (R*v)+(C*m)/(v+m)
```

```python
movies_data.sort_values(by='Weighted_Average',ascending=False).head(5)
```

```python
movies_data['popularity'].head()
```

movies_data.columns

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
finalresultant = scaling.fit_transform(movies_data[['popularity','Weighted_Average']])
normalized_df = pd.DataFrame(finalresultant,columns=['popularity','Weighted_Average'])
movies_data[['popularity','Weighted_Average']] = normalized_df
```

```python
movies_data[['popularity','Weighted_Average']].head(5)
```

```python
movies_data['score'] = movies_data['popularity']*0.5+movies_data['Weighted_Average']*0.5
```

```python
movies_data = movies_data.sort_values(['score'],ascending=False)
```
